## **Ejercisio de Regresion LINEAL**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
df = pd.read_csv('ingreso.csv')
df
```

```python
plt.figure() #Creamos el espacio de la figura

plt.scatter(df['horas'], df['ingreso']) #Graficamos los datos
plt.title('Horas vs ingresos')
plt.show()
```

```python
x = df['horas'].values.reshape(-1,1)
y = df['ingreso'].values.reshape(-1,1)

x
y
```

```python
from sklearn.preprocessing import StandardScaler

# instanciamos el modelo de escalamiento
sc_x = StandardScaler()
sc_y = StandardScaler()
```

```python
# estandarizamos los datos z=(x-u)/std
x_standard = sc_x.fit_transform(x)
y_standard = sc_y.fit_transform(y)
```

```python
from sklearn.linear_model import LinearRegression

# Instanciamos y ajustamos el modelo de regresi√≥n lineal
slr = LinearRegression() # Standar Linear Regression (slr)
slr.fit(x_standard,y_standard)
```

```python
plt.scatter(x_standard, y_standard)
plt.plot(x_standard, slr.predict(x_standard), color='red')
```
