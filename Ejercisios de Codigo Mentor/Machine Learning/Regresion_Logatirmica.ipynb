```python
#Importación de librerías
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Cargar datos
X = np.arange (1, 16, 1)
y = 3 + np.log10(X)

print(X)
print(y)
```

```python
plt.scatter(X, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresión Logaritmica')
plt.show()
```

```python
def fit(x, y):
    fit = np.polyfit(np.log(x), y, 1)
    a = fit[1]
    b = fit[0]
    return a, b
```

```python
a, b = fit(X, y)
print(a, b)
```

```python
def predict(a, b, x):
    y = a + b * np.log(x)
    return y
```

```python
y_pred = predict(a, b, X)
print(y_pred)
```

```python
plt.plot(X, y_pred, color = 'red')
plt.scatter(X, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresión Logaritmica')
plt.show()
```
