- Crear 20 filas minimo
- Graficar X1 Vs Y, X2 Vs Y.
- Realizar Regresión Lineal Polinomica.
- Hallar coeficientes e interceptos.
- Imprimir en pantalla la ecuación de resultado

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import random
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.DataFrame()

df['X1'] = [random.randint(1, 10) for _ in range(20)]
df['Y1'] = 0.125*df['X1']**3 + df['X1']**2 + 5
df['Y2'] = np.random.normal(0, 1, 20) + df['Y1']

df
```

```python
plt.figure()

plt.scatter(df['X1'], df['Y1'], color='red', marker='x')
plt.xlabel('X1')
plt.ylabel('Y1')
plt.title('X1 VS Y1')
plt.show()

plt.scatter(df['X1'], df['Y2'], color='blue', marker='x')
plt.xlabel('X1')
plt.ylabel('Y2')
plt.title('X2 VS Y2')
plt.show()
```

```python
#Generamos el array de nuestra variable
X = df['X1'].values

#Creamos el dato de tipo polinomico de grado 2, con el PolynomialFeatures, donde definimos el grado
poly_regressor = PolynomialFeatures(degree=3)

#Pasamos nuestros datos tipo vector columna para que el regresor cree la matriz de los datos
poly_feature = poly_regressor.fit_transform(X.reshape(-1, 1))

#imprimimos los valores de cabecera de x
print(poly_regressor.get_feature_names_out())
#imprimimos los valores de la matriz
print(poly_feature)
```

```python
#Creo el modelo de regresion y pasamos el Y como array
model = LinearRegression()
#Hace la creacion del modelo
model.fit(poly_feature, df['Y1'].values)

#Crea las prediciones
y_pred = model.predict(poly_feature)

#Imprime los coeficientes y el valor del interceptor
print(model.coef_)
print(model.intercept_)
```

```python
plt.figure()

plt.scatter(df['X1'], df['Y1'], color='red', marker='x')
plt.plot(X, y_pred, color='blue')
plt.xlabel('X1')
plt.ylabel('Y')
plt.title('X1 VS Y')
plt.show()
```
