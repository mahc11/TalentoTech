```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame()

df['X1'] = [random.randint(1, 100) for _ in range(20)]
df['X2'] = [random.randint(1, 100) for _ in range(20)]
df['Y'] = df['X1']*2 + df['X2']*3 + np.random.randn(20)*5

df
```

```python
plt.figure()

plt.scatter(df['X1'], df['Y'], color='red', marker='x')
plt.xlabel('X1')
plt.ylabel('Y')
plt.title('X1 VS Y')
plt.show()

plt.scatter(df['X2'], df['Y'], color='blue', marker='x')
plt.xlabel('X2')
plt.ylabel('Y')
plt.title('X2 VS Y')
plt.show()
```

```python
model = LinearRegression()

X = df[['X1', 'X2']]
y = df['Y']

model.fit(X, y)
```

```python
print('Coefficientes: ', model.coef_)
print('intercept: ', model.intercept_)
print('R2', model.score(X, y))

y_pred = model.predict(X)
```
